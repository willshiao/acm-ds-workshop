```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from pandas import DataFrame
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
from IPython.display import clear_output
from sklearn.cluster import KMeans
import time
%matplotlib inline
```

```python
df = pd.read_table('data/turtles.dat')
```

```python
colors = ['red', 'blue']
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

n = 100

ax.scatter(df['length'], df['breadth'], df['height'], c=df['gender'], cmap=ListedColormap(colors))
ax.set_xlabel('Turtle Length')
ax.set_ylabel('Turtle Breadth')
ax.set_zlabel('Turtle Height')

ax.view_init(25, 120)
plt.draw()
```

```python
colors = ['red', 'blue']
def plot_kmeans(clusters, angle = 120):
    data = np.array([df['length'], df['breadth'], df['height'], df['gender']]).T
    X = data[:, :-1]

    mdl = KMeans(n_clusters=clusters)
    y = mdl.fit_predict(X)
    centers = mdl.cluster_centers_

    fig = plt.figure(figsize=(14, 10))
    ax = fig.add_subplot(111, projection='3d')

    turt = ax.scatter(df['length'], df['breadth'], df['height'], c=y, cmap=ListedColormap(colors))
    ax.set_xlabel('Turtle Length')
    ax.set_ylabel('Turtle Breadth')
    ax.set_zlabel('Turtle Height')

    ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], marker=11, s=500)

    ax.view_init(25, angle)
    plt.show()

plot_kmeans(2, 180)
```

```python
plot_kmeans(4)
```

```python
plot_kmeans(10)
```
